```python
import pandas as pd
```

```python
df=pd.read_csv(r'C:\Users\Muthu\VS\orders.csv')
df
```

```python
df.isnull().sum()
```

```python
# filling null values
df = df.fillna(value = 0)
df
```

```python
df.isnull().sum()
```

```python
# columns rename
df = df.rename(columns={
    'Order Id': 'order_id',
    'Ship Mode': 'ship_mode',
    'Order Date': 'order_date',
    'Postal Code': 'postal_code',
    'Sub Category': 'sub_category',
    'Product Id': 'product_id',
    'cost price': 'cost_price',
    'List Price': 'list_price',
    'Discount Percent': 'discount_percent'
})
df.columns = df.columns.str.lower()
df
```

```python
df
```

```python
# adding 3 columns
df['discount']=df['list_price']*df['discount_percent']*.01
df['sale_price']=df['list_price']-df['discount']
df['profit']=df['sale_price']-df['cost_price']
df
```

```python
# table trimming
df = df.applymap(lambda x: x.rstrip() if isinstance(x, str) else x)

print("\nAfter trimming:")
df
```

```python
# Table two split
df1=df[['order_id','order_date','ship_mode','country','state','city','postal_code','region','category','sub_category']] 
df1
```

```python
df2=df[['segment','product_id','cost_price','list_price','quantity','discount_percent','discount','sale_price','profit']]
df2
```

```python
df1=df[['order_id','order_date','ship_mode','country','state','city','postal_code','region','category','sub_category']] 
df1
```

```python
mycursor.execute("use retail_data")
```

```python
mycursor.execute("""
CREATE TABLE order_1
(
    order_id int PRIMARY KEY,         
    order_date DATE NOT NULL,                
    ship_mode VARCHAR(50),                   
    country VARCHAR(100),                   
    state VARCHAR(100),                     
    city VARCHAR(100),                     
    postal_code int,                
    region VARCHAR(50),                    
    category VARCHAR(50),                
    sub_category VARCHAR(50)              
);
""")
print("Table Created successfully")
```

```python
import mysql.connector

# Assuming df1 is a pandas DataFrame

insert_query = """
    INSERT INTO order_1 (
        order_id,
        order_date,
        ship_mode,
        country,
        state,
        city,
        postal_code,
        region,
        category,
        sub_category
    ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

data = df1.values.tolist()

try:
    mycursor = connection.cursor()
    mycursor.executemany(insert_query, data)
    connection.commit()
    print("Data inserted successfully.")
except mysql.connector.Error as err:
    print(f"Error: {err}")
    if connection.is_connected():
        connection.rollback()
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    if connection.is_connected():
        connection.rollback()
finally:
    if 'mycursor' in locals() and mycursor is not None:
        mycursor.close()
```
